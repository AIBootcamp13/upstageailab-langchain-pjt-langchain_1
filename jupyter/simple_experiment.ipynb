### Test

```python
# 기본 설정으로 초기화 - config_name은 파일명만 사용
qa = LegalQA(config_name="config")

# 테스트 질문
question = "업무상 재해로 인한 정신질환 치료가 산재보험 적용을 받을 수 있는 조건은 뭐야?"
result = qa.ask(question)

print(f"질문: {question}")
print(f"\n답변: {result.get('answer', result.get('error'))}")
```
